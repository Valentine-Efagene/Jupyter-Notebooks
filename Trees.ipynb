```python
class TreeNode:
    right = None
    left = None
    parent = None
    key = None
    
    def __init__(self, key, r = None, l = None, p = None):
        self.right = r
        self.left = l
        self.parent = p
        self.key = key
        
    def setLeft(self, l):
        self.left = l
        
        if (l != None):
            l.parent = self
        
    def setRight(self, r):
        self.right = r
        
        if (r != None):
            r.parent = self
        
    def setParent(self, p):
        self.parent = p
        
    def inOrderTreeWalk(self):
        if (self != None):
            yield from inOrderTreeWalk(self.left)
        
            if(self.key != None):
                yield self.key
        
            yield from inOrderTreeWalk(self.right)
            
    def search(self, key):
        x = self
        
        while(x != None and key != x.key):
            if (key < x.key):
                x = x.left
            else:
                x = x.right
                
        return x
        
    def root(self):
        root = self
        
        while (True):
            root = root.parent
            
            if (root.parent == None):
                return root
            
    def min(self):
        x = self
        
        while(x.left != None):
            x = x.left
    
        return x
    
    def max(self):
        x = self
        
        while(x.right != None):
            x = x.right
    
        return x
    
    def successor(self):
        x = self
        
        if(x.right != None):
            return x.right.min()
        
        y = x.parent
        
        while (y != None and x == y.right):
            x = y
            y = y.parent
            
        return y
```

```python
n6 = TreeNode(6)
n5 = TreeNode(5)
n5r = TreeNode(5)
n2 = TreeNode(2)
n7 = TreeNode(7)
n8 = TreeNode(8)

n6.setLeft(n5)
n6.setRight(n7)
n5.setLeft(n2)
n5.setRight(n5r)
n7.setRight(n8)
```

## Walk

### In-order Walk

```python
def inOrderTreeWalk(x):
    if (x != None):
        yield from inOrderTreeWalk(x.left)
        
        if (x.key != None):
            yield x.key
        
        yield from inOrderTreeWalk(x.right)
```

## Visualize

```python
from graphviz import Graph

def visualize(node):
    g = Graph(node_attr={'color': 'lightgrey', 'shape': 'circle', 'style': 'filled'}, graph_attr = {'splines': 'polyline'})
    drawOne(g, node)
    return g
    
def drawOne(g, n):
    g.node(str(n), str(n.key))
    
    if (n.left) and (n.left.key):
        g.node(str(n.left), str(n.left.key))
        g.edge(str(n), str(n.left))
        drawOne(g, n.left)
        
    if (n.right) and (n.right.key):
        g.node(str(n.right), str(n.right.key))
        g.edge(str(n), str(n.right))
        drawOne(g, n.right)    
```

```python
visualize(n6)
```

## Query

### Search

```python
def search(x, key):
    if x == None or key == x.key:
        return x
    
    if key < x.key:
        return search(x.left, key)
    else:
        return search(x.right, key)
```
